```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('PDA_Dados_Cursos_Graduacao_Brasil.csv.crdownload')

organizacao = df['ORGANIZACAO_ACADEMICA'].value_counts().nlargest(3)
print(organizacao)

organizacao.plot(kind='pie', autopct='%1.1f%%')
plt.title('Organizações')
plt.show()

# Exibição dos cursos 

curso = df['NOME_CURSO'].value_counts().nlargest(10)
print(curso)

curso.plot(kind='pie', autopct='%1.1f%%')
plt.title('Cursos')
plt.show()

# Grafico barras, exibição de estados

uf = df['UF'].value_counts().nlargest(10)
print(uf)

uf.plot(kind='bar')
plt.xlabel('Estados')
plt.ylabel('Total vagas autorizadas')
plt.title('Estados com maior indice de graduação')
plt.show()
```
